```python
!pip install yfinance
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Apply a professional style
plt.style.use('fivethirtyeight')

# 2. DATA: Pull live data for your Tech Giants
tickers = ["AAPL", "TSLA", "NVDA"]
# We pull 'Close' prices for the last month
df = yf.download(tickers, period="1mo")['Close']

# Fix for the 2026 'Multi-Index' format (Ensures the labels are clean)
if isinstance(df.columns, pd.MultiIndex):
    df.columns = df.columns.get_level_values(0)

# 3. ANALYSIS: Calculate Daily Returns to see performance
# This normalizes the data so we can see who is growing the fastest
returns = (df / df.iloc[0] - 1) * 100

# 4. VISUALIZATION: Create the professional chart
plt.figure(figsize=(12, 6))
for column in returns.columns:
    plt.plot(returns.index, returns[column], label=column, linewidth=2)

# Add titles and labels
plt.title('Market Performance: AAPL vs. TSLA vs. NVDA (Last 30 Days)', fontsize=14)
plt.ylabel('Percentage Growth (%)')
plt.xlabel('Date')
plt.legend(loc='upper left')
plt.axhline(0, color='black', lw=1, ls='--') # Add a "Break-even" line
plt.grid(True, alpha=0.3)

# 5. SHOW THE RESULT
plt.tight_layout()
plt.show()

# 6. SUMMARY TABLE
print("\n--- PERFORMANCE SUMMARY ---")
print(returns.tail(1).T.rename(columns={returns.index[-1]: 'Total Growth %'}))
```
